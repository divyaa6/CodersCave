***Visualizations from given dataset***

```python
plt.figure(figsize=(12,6))
plt.plot(ddf["DATE"],ddf['DailyAverageDryBulbTemperature'],label='Varaition of Daily Temperature')
plt.ylabel("Temperature")
plt.xlabel("Date")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(ddf["DATE"],ddf['DailyAverageRelativeHumidity'],label='Varaition of Daily Average Relative Humidity',color='green')
plt.ylabel("Relative Humidity")
plt.xlabel("Date")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(ddf["DATE"],ddf['DailyPrecipitation'],label='Varaition of Daily Precipitation')
plt.ylabel("Precipitation")
plt.xlabel("Date")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(ddf["DailyPeakWindDirection"],density=True,bins=31,label='Daily peak Wind Direction Trends')
plt.ylabel("Density")
plt.xlabel("Wind Direction")
plt.legend()
plt.show()
```

```python
tm=pd.to_datetime(ddf['DATE'])
tm
```

```python
ddf.groupby(tm)['DailyAverageDryBulbTemperature'].mean().plot()
```

```python
ddf['Season'] = (tm.dt.month % 12 + 3) // 3
seasonal_temperature = ddf.groupby('Season')['DailyAverageDryBulbTemperature'].mean()

plt.figure(figsize=(12, 6))
plt.plot(seasonal_temperature.index, seasonal_temperature.values, marker='o')
plt.title('Seasonal Average Temperature Trends')
plt.xlabel('Season')
plt.ylabel('Average Temperature (°F)')
plt.xticks([1, 2, 3, 4], ['Winter', 'Spring', 'Summer', 'Fall'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(ddf["Sunset"],ddf['Sunset'],label='Sunrise Vs Sunset')
plt.ylabel("Sunset")
plt.xlabel("Sunrise")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(ddf['DailySnowDepth'],label='Variation of daily snow depth')
plt.legend()
plt.show()
```
